# Encrypt your passwords with ease!

Encrypt your passwords into a local file!


**How it works:**
1. Select an option
2. Enter the website
3. Enter the username
4. Enter the password
5. Wait for the password to be encrypted/decrypted


**Required modules:**
- `pip install cryptography`



```python
import csv
import os
from cryptography.fernet import Fernet

passwords = []
key = Fernet.generate_key()
cipher_suite = Fernet(key)

def encrypt_password(password):
    return cipher_suite.encrypt(password.encode())

def decrypt_password(encrypted_password):
    return cipher_suite.decrypt(encrypted_password).decode()

def add_password():
    website = input("Website: ")
    username = input("Username: ")
    password = input("Password: ")
    encrypted_password = encrypt_password(password)
    passwords.append({
        "website": website,
        "username": username,
        "password": encrypted_password
    })
    with open('passwords.csv', mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([website, username, encrypted_password])

def get_password(website):
    for entry in passwords:
        if entry["website"] == website:
            username = entry["username"]
            encrypted_password = entry["password"]
            decrypted_password = decrypt_password(encrypted_password)        
            print(f"Website: {website}")
            print(f"Username: {username}")
            print(f"Password: {decrypted_password}")
            return
    print("Website not found")

# Check if passwords.csv exists, create it if it doesn't
if not os.path.exists('passwords.csv'):
    print("üìÅ passwords.csv file not found. Creating a new one...")
    with open('passwords.csv', mode='w', newline='') as file:
        pass  # Create empty file
    print("‚úÖ passwords.csv file created successfully!")
else:
    print("üìÅ Found existing passwords.csv file.")

# Load existing passwords from CSV
try:
    with open('passwords.csv', mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            if row:  # Skip empty rows
                passwords.append({
                    "website": row[0],
                    "username": row[1],
                    "password": row[2]
                })
    if passwords:
        print(f"üìã Loaded {len(passwords)} existing password(s) from file.")
    else:
        print("üìã No existing passwords found in file.")
except Exception as e:
    print(f"‚ö†Ô∏è Error reading passwords.csv: {e}")

while True:
    print("\n1. Add Password")
    print("2. Get Password")
    print("3. Exit")
    
    choice = input("Enter your choice: ")
    
    if choice == '1':
        add_password()
    elif choice == '2':
        website = input("Enter website: ")
        get_password(website)
    elif choice == '3':
        break
    else:
        print("Invalid choice")
```
